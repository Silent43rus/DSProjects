```python
from scipy.stats import linregress
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
```

```python
tab_res1 = pd.read_excel('C:/DataPreprocessing_spc500_01012022_20012022/result_1_01012022_20012022.xlsx')
tab_res1 = tab_res1.drop(columns = ['energycount','Unnamed: 0'], axis = 1)
```

```python
a = tab_res1['realtime']
```

```python
import matplotlib.pyplot as plt
def cov(x, y):
    assert x.size == y.size
    return ((x - x.mean()) * (y - y.mean())).sum()/(x.size - 1)

def cor(x, y):
    return cov(x, y)/(np.std(x, ddof=1)*np.std(y, ddof=1))


fig, ax1  = plt.subplots(1, 1, figsize=(16, 3))
ax1.scatter(tab_res1['perfcount'], tab_res1['gazcount'])
plt.show()
print('cor: ', cor(tab_res1['perfcount'], tab_res1['gazcount']))
```

```python
coef = np.polyfit(tab_res1['perfcount'],tab_res1['gazcount'], 1)
poly1d_fn = np.poly1d(coef)

plt.plot(tab_res1['perfcount'],tab_res1['gazcount'], 'yo', tab_res1['perfcount'], poly1d_fn(tab_res1['perfcount']), '--k')

print(linregress(tab_res1['perfcount'], tab_res1['gazcount']))
plt.show()
```

```python
b1 = tab_res1['gazcount'].std()/tab_res1['perfcount'].std()*cor(tab_res1['perfcount'], tab_res1['gazcount'])
b0 = tab_res1['gazcount'].mean() - b1*tab_res1['perfcount'].mean()
f = lambda x: b0 + b1*x
y_pred = f(tab_res1['perfcount'])
plt.scatter(tab_res1['perfcount'], tab_res1['gazcount'])
plt.plot(tab_res1['perfcount'], y_pred, color='r')
plt.show()
```
